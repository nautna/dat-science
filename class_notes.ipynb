```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import sklearn.datasets

#1. Load DataSet
iris_dataset = sklearn.datasets.load_iris()

X = iris_dataset.data
y = iris_dataset.target

#labels
print "Target classes", iris_dataset.target_names
print "Feature names", iris_dataset.feature_names

#actual data
print iris_dataset.data
print iris_dataset.target
```

```python
from sklearn.neighbors import KNeighborsClassifier #classifier
from sklearn.cross_validation import train_test_split #train-test
```

```python
#2. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
print len(X_train) #sanity check to make sure that syntax above is correct
print len(y_train)
```

```python
#3. Define model/params
k = 5
knn_classif = KNeighborsClassifier(k)

#4. fit model
knn_classif.fit(X_train, y_train)

#5. score
knn_classif.score(X_test, y_test)
```

```python
def knn_script(k):
    #3. Define model/params
    knn_classif = KNeighborsClassifier(k)

    #4. fit model
    knn_classif.fit(X_train, y_train)

    #5. score
    return knn_classif.score(X_test, y_test)
    
```

```python
for i in range(2, 100):
    print "k: %s => %0.3f" % (i, knn_script(i))

scores = [knn_script(i) for i in range(2, 100)]
```

```python
from sklearn.grid_search import GridSearchCV
#defines, fit and scores models

#minimum of three input for cross validation
# 1. Define the classifier

classifier = KNeighborsClassifier()

# 2. Define the parameter space
parameter_space = {
    'n_neighbors': range(2, 99),
    'weights': ['uniform', 'distance']
}

#3. Define how to validate the model
cross_validator = KFold(len(X_iris), 20)

#GridSearchCV(model, range of parameters, scoring)
gridsearch = GridSearchCV(estimator=classifier,
                          param_grid=parameter_space,
                          cv=cross_validator)

# fit is a common function in SKLearn
gridsearch.fit(X_iris, y_iris)

# underscore item is a 'results' object; it is calculated
# gridsearch.best_score_

print "Best score was %0.3f" % gridsearch.best_score_
print "Best params was: %s" % gridsearch.best_params_
```
